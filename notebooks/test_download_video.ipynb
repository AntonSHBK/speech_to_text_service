```python
from pathlib import Path

import pandas as pd

pd.set_option('display.max_colwidth', 500) 
```

```python
import sys
import os

project_path = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(project_path)
```

```python
DATA_DIR = Path(project_path) / "data"
DATA_AUDIO_DIR = DATA_DIR / "audio"
```

```python
VIDEO_URL_1 = "https://www.youtube.com/watch?v=LPZh9BOjkQs"
RUTUBE_URL_2 = "https://rutube.ru/video/b764735bfd9eb076d7232451770c4ef5/"

```

```python
from pathlib import Path
from yt_dlp import YoutubeDL


def download_youtube_audio(url: str, output_dir: Path) -> Path:
    output_dir.mkdir(parents=True, exist_ok=True)

    ydl_opts = {
        "format": "bestaudio/best",
        "outtmpl": str(output_dir / "%(id)s.%(ext)s"),
        "postprocessors": [
            {
                "key": "FFmpegExtractAudio",
                "preferredcodec": "wav",
                "preferredquality": "0",
            }
        ],
        "quiet": True,
        "noplaylist": True,
    }

    with YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=True)
        audio_path = output_dir / f"{info['id']}.wav"
        return audio_path


```

```python
audio_file = download_youtube_audio(
    url=VIDEO_URL_1,
    output_dir=DATA_AUDIO_DIR
)

print(f"Аудио скачано: {audio_file}")
```

```python
audio_file = download_youtube_audio(
    url=RUTUBE_URL_2,
    output_dir=DATA_AUDIO_DIR
)

print(f"Аудио скачано: {audio_file}")
```
